```python
import json
import numpy as np
import string
from random import choice, sample
```

```python
nombres   = ['Victoria','Renata','Sofia','Valeria','María','Fernanda','Valentina','Ximena','Regina','Camila','Santiago','Daniel','Miguel','David','Leonardo','Diego','Emiliano','Juan','Fernando','Luis','Francisco','Carlos']
apellidos = ['Perez','Arcos','Delgadillo','Caro','Correa','Zapata','Espinoza','Sanchez','Toledo', 'Delgado', 'Herrera','Bravo','Portillo','Chavez','Tejeda','Hernandez','Lozano','Jimenez','Torres','Salgado','Salvador','Campos']
materias  = ['Historia', 'Matematicas', 'Geografia', 'Machine Learning', 'Patrones de Diseño', 'Japones', 'Redaccion','Bases de Datos']
carreras  = ['Inteligencia Artificial','Diseño','Electronica','Industrial']
ciudades  = ['Aguascalientes', 'Zacatecas', 'CDMX', 'Jalisco']
mails_fin = ['outlook.com','gmail.com','hotmail.com']

alumnos      = []
profesores   = []
supervisores = []

calf_vals = []
for i in range(7): calf_vals.append(10)
for i in range(9): calf_vals.append(9)
for i in range(5): calf_vals.append(8)
calf_vals.extend([7,7,7,6,6,5])

json_dict = dict()
```

```python
def getPassword(length):
    s = ''.join(choice(string.ascii_letters) for i in range(length))
    return s

def getMail(a, b):
    fin = mails_fin[np.random.randint(len(mails_fin))]
    return a.lower()+'.'+b.lower()+'@'+fin

def getName():
    nb, ap1, ap2 = np.random.randint(len(nombres)), np.random.randint(len(apellidos)), np.random.randint(len(apellidos))
    a, b, c = nombres[nb], apellidos[ap1], apellidos[ap2]
    return str(a+' '+b+' '+c), getMail(a,b)
    
```

```python
ids = []
for i in range(50):
    name, mail = getName()
    calf = [ calf_vals[np.random.randint(len(calf_vals))] for _ in range(len(materias))]
    año = 1999 + np.random.randint(4)
    password = getPassword(6)
    ciudad = ciudades[np.random.randint(len(ciudades))]
    carrera = carreras[np.random.randint(len(carreras))]
    ids.append(i)
    alumnos.append({'Id':i ,'Nombre':name, 'Mail':mail, 'FechaDeNacimiento':año, 'Calificaciones':calf, 'Ciudad':ciudad, 'Carrera':carrera, 'Password':password})
    
```

```python
for i in range(8):
    name, mail = getName()
    año = 1988 + np.random.randint(15)
    ciudad = ciudades[np.random.randint(len(ciudades))]
    alumnos_profe = sample(ids, 10)
    alumnos_profe.sort()
    password = getPassword(6)
    profesores.append( { 'Id':i, 'Nombre':name, 'Mail':mail, 'FechaDeNacimiento':año, 'Ciudad':ciudad, 'Alumnos':alumnos_profe , 'Password':password })
```

```python
for i in range(3):
    name, mail = getName()
    año = 1985 + np.random.randint(20)
    ciudad = ciudades[np.random.randint(len(ciudades))]
    password = getPassword(6)
    supervisores.append( { 'Id':i, 'Nombre':name, 'Mail':mail, 'FechaDeNacimiento':año, 'Ciudad':ciudad, 'Password':password} )
```

```python
def debugueo():
    calf = [ calf_vals[np.random.randint(len(calf_vals))] for _ in range(len(materias))]
    password = getPassword(6)
    ciudad = ciudades[np.random.randint(len(ciudades))]
    carrera = carreras[np.random.randint(len(carreras))]
    alumnos.append({'Id':-1 ,'Nombre':'juan', 'Mail':'a', 'FechaDeNacimiento':2022, 'Calificaciones':calf, 'Ciudad':ciudad, 'Carrera':carrera, 'Password':'a'})

debugueo()

json_dict = {
    'Alumnos' : alumnos,
    'Profesores' : profesores,
    'Supervisores' : supervisores,
    'Materias': materias
}

with open('Examen2doParcial/json_data.json', 'w') as outfile:
    json.dump(json_dict, outfile)
```
