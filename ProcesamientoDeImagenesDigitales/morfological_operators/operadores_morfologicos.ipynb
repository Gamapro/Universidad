```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import numpy.ma as ma
from PIL import Image
```

```python
img = Image.open('../images/huella2.png').convert('L')
img = np.int16(img)
plt.axis('off')
plt.imshow(img, cmap="gray")
```

```python
# Apply a open operation
kernel = np.ones((7,7),np.uint8)
# Set image to 0 and 1
img = np.where(img > 0, 1, 0)
# Erosion
img2 = np.copy(img)
for i in range(3, img.shape[0]-3):
    for j in range(3, img.shape[1]-3):
        cont = np.sum(img[i-3:i+4, j-3:j+4] * kernel)
        total = np.sum(kernel) 
        if cont != total:
            img2[i, j] = 0
plt.axis('off')
plt.imshow(img2, cmap="gray")
```

```python
# Dilation
opening = np.copy(img)
for i in range(3, img.shape[0]-3):
    for j in range(3, img.shape[1]-3):
        cont = np.sum(img2[i-3:i+4, j-3:j+4])
        if cont == 0:
            opening[i, j] = 0
# Show images
plt.axis('off')
plt.imshow(opening, cmap="gray")
```
