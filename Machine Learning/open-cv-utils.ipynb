```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import cv2
```

```python
img_BGR = cv2.imread('images/mariposa.jpg')
img = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)  
hsv = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2HSV) 
gray = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2GRAY) 
plt.imshow(img)
print(img.shape)
```

```python
x,y,ch = img.shape

a, b = np.array([90, 50, 50]), np.array([130, 255,255])
mask_range = cv2.inRange(hsv, a, b)

img_mask = cv2.bitwise_and(img, img, mask=mask_range)
```

```python
fig, ax = plt.subplots(2,3,figsize=(16,8))

ax[0,0].imshow(img)
ax[0,1].imshow(hsv)
ax[0,2].imshow(gray, cmap='gray')
ax[1,0].imshow(img_mask)
ax[1,1].imshow(gray, cmap='inferno')
ax[1,2].imshow(img_BGR)
```
