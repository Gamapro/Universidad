```python
from pynput.keyboard import Key,Controller
import time
import serial
```

```python
def read():
    data = ser.readline()
    data = data.decode('UTF-8').split()  
    return data
def direccion_lateral(val):
    if val<-.2:
        keyboard.release('d')
        keyboard.press('a')
        return "izquierda"

    elif val>.2:
        keyboard.release('a')
        keyboard.press('d')
        return "derecha"
    else:
        keyboard.release('a')
        keyboard.release('d')
        return "quieto"
def direccion_frontal(val):
    if val<-.2:
        keyboard.release('s')
        keyboard.press('w')
        return "enfrente"
    elif val>.2:
        keyboard.press('s')
        keyboard.release('w')
        return "atras"
    else:
        keyboard.release('s')
        keyboard.release('w')
        return "quieto"
def salto(val):
    if val > 650:
        keyboard.press(' ')
      #  keyboard.release('r')
        return "salto"
    else:  
        keyboard.release(' ')      
        "parado"
```

```python
time.sleep(4)
keyboard = Controller()
with serial.Serial() as ser:
    ser.baudrate = 9600
    ser.port = 'COM8'
    ser.open()    
    while True:
        vals = read()
        vals = list(map(float, vals))
        x = (direccion_frontal(vals[1]),vals[1],direccion_lateral(vals[2]),vals[2],salto(vals[0]),vals[0])
        print(x)
```
